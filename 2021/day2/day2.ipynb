# Puzzle 1

```python
import pandas as pd

df = pd.read_csv("input.txt", header=None, sep=" ")
df.columns = ["Direction", "Amount"]

df["delta_x"] = (
    df["Direction"].map({"forward": 1, "down": 0, "up": 0}) * df["Amount"]
)
df["delta_y"] = (
    df["Direction"].map({"forward": 0, "down": 1, "up": -1}) * df["Amount"]
)
df["x"] = df.delta_x.cumsum()
df["y"] = df.delta_y.cumsum()
df.loc[-1] = {"x": 0, "y": 0}
df = df.sort_index()

df
```
