# Puzzle 1

```python
import pandas as pd
from IPython.display import display

df = pd.read_csv("input.txt", header=None, names=["bin"], dtype={"bin": str})
df = pd.concat([df, df.bin.str.split("", expand=True)], axis=1)
df = df.drop(columns=[0, 13])
# df.columns = ["bin"] + [str(x) for x in range(12)]
df
```

```python
gamma_bin = df.drop("bin", axis=1).mode().squeeze().str.cat()
gamma = int(gamma_bin, 2)
epsilon_bin = (
    df.drop("bin", axis=1)
    .apply(pd.value_counts)
    .apply(lambda x: x.idxmin())
    .squeeze()
    .str.cat()
)
epsilon = int(epsilon_bin, 2)

print(gamma)
print(epsilon)
```

```python
epsilon * gamma
```

```python
df_ogr = df.copy()
i = 1
while df_ogr.shape[0]>1:
    vc = df_ogr[i].value_counts()
    if vc[0]==vc[1]:
        digit = "1"
    else:
        digit = vc.index[0][0]
    df_ogr = df_ogr[df_ogr[i] == digit]
    # print(i)
    # display(vc)
    # display(digit)
    # display(df_ogr)
    i += 1
ogr_bin = df_ogr.drop("bin", axis=1).mode().squeeze().str.cat()
ogr = int(ogr_bin, 2)
ogr
```

```python
df_csr = df.copy()
i = 1
while df_csr.shape[0]>1:
    vc = df_csr[i].value_counts()
    if vc[0]==vc[1]:
        digit = "0"
    else:
        digit = vc.index[-1][0]
    df_csr = df_csr[df_csr[i] == digit]
    # print(i)
    # display(vc)
    # display(digit)
    # display(df_csr)
    i += 1
csr_bin = df_csr.drop("bin", axis=1).mode().squeeze().str.cat()
csr = int(csr_bin, 2)
csr
```

```python
ogr * csr
```
