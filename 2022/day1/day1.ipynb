# Read Input

```python
with open("input.txt", "r") as f:
    input = f.read()
input
```

```python
elves = [[int(y) for y in x.split("\n")  if y != ""] for x in input.split("\n\n")]
elves
```

# Part 1

```python
answer1 = max(sum(elf) for elf in elves)
answer1
```

# Part 2

```python
calorie_list = sorted([sum(elf) for elf in elves], reverse=True)
answer2 = sum(calorie_list[:3])
answer2
```
